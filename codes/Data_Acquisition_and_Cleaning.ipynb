```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set seeds for reproducibility
np.random.seed(42)
plt.style.use('seaborn-v0_8-whitegrid')
print("Execution Environment Initialized.")
```

```python
# Load the TASI dataset sourced from Kaggle
# Ensure the CSV file is placed in the '../data/' directory
df = pd.read_csv('historical_data.csv')
df['date'] = pd.to_datetime(df['date'])

# Sort data by ticker and date to preserve time-series sequence
df = df.sort_values(['ticker', 'date']).reset_index(drop=True)
print(f"Dataset Loaded Successfully: {df.shape[0]} observations.")
```

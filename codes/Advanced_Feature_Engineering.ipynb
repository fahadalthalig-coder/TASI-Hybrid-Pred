```python
def generate_advanced_features(group):
    """
    Generates technical indicators and the shifted target variable.
    Target: Close price shifted by 18 trading days.
    """
    res = pd.DataFrame(index=group.index)

    # Forecasting Horizon: 18 Trading Days (~3.5 weeks)
    res['target'] = group['close'].shift(-18)

    # Structural Anchor: 200-Day Moving Average (MA200)
    res['ma_200'] = group['close'].rolling(window=200).mean()

    # Momentum Proxy: Relative Strength Index (RSI) components
    res['rsi'] = group['close'].diff().rolling(14).mean()

    return res
```

```python
# Apply feature engineering grouped by ticker
data = df.groupby('ticker').apply(generate_advanced_features).dropna().reset_index(drop=True)
print("Feature Engineering Complete: 18-day prediction horizon established.")
```
